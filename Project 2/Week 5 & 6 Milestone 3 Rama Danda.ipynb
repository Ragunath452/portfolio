```python
#Replace Headers
df=df.rename(columns={'Manufacturer':'MANUFACTURER', 'Make':'MAKE', 'Model':'MODEL', 'State':'STATE', 'Facility':'FACILITY'})
```

```python
#Format data into a more readable format
#Remove extra '[value]' in MANUFACTURER

df['MANUFACTURER']=df['MANUFACTURER'].apply(lambda x: x.split('[')[0])

```
